```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [0,1,2]
y = [100,200,300]
```

```python
plt.plot(x,y)
plt.show()
```

```python
housing = pd.DataFrame({'rooms':[1,1,2,2,2,3,3,3],
                       'price':[100,120,190,200,230,310,330,305]})
housing
```

```python
plt.scatter(housing['rooms'],housing['price'])
```

```python
plt.plot(x,y)
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

```python
plt.plot(x,y)

# Axis and ticks
plt.xlim(0,2)
plt.ylim(100,300)


# Labeling
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

```python
plt.plot(x,y,color='red')

# Axis and ticks
plt.xlim(0,2)
plt.ylim(100,300)


# Labeling
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

```python
plt.plot(x,y,color='red',marker='o',markersize=20,linestyle='--')

# Axis and ticks
plt.xlim(0,2)
plt.ylim(100,300)


# Labeling
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```
